# Data Standardization and Preproccessing in Python

```python
# import the libraries
import pandas as pd
import numpy as np
import sklearn.datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
# Loading the dataset
dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(dataset)
```

```python
# loading the dataset into a pandas dataframe
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
```

```python
df.head()
```

```python
df.shape
```

```python
X = df
y = dataset.target
```

```python
print(X)
```

```python
print(y)
```

```python
# Splitting the dataset in training and testing set
X_train,X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state=3, stratify=y)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Standardinzing the Data

```python
print(dataset.data.std())
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
X_train_starndardized = scaler.transform(X_train)
print(X_train_starndardized)
```

```python
print(X_train_starndardized.std())
```

```python
X_test_standardized = scaler.transform(X_test)
print(X_test_standardized)
```

```python
print(X_test_standardized.std())
```
