# Building Your First Machine Learning Regression Model

# Importing the Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Importing the Dataset

```python
# Importing the Dataset
sales_df = pd.read_csv("advertising_and_sales_.csv")
```

```python
# Checking for the first 5 rows of the data
sales_df.head()
```

Preprocessing (X and y)
Code:



Explanation:
X (Features):
-------

This is the input (the "radio" column). It is a 2D matrix (even with one column) because models expect a matrix of inputs.

Explanation
y (Target):
-------

This is the output (the "sales" column). It is a 1D vector because it is just the answer we want.

```python
# Feature : radio column
X = sales_df["radio"].values
# Target: sales column
y = sales_df['sales'].values
```

```python
print(X)
```

```python
print(y)
```

# Reshaping X- So that scikit-learn sees it as a 2D

```python
# Reshape X for scikit-learn (must be 2D)
X = X.reshape(-1, 1)
print("X shape:", X.shape)
print("y shape:", y.shape)
```


# Importing the  Linear Regression Model


```python
# Import The Linear_Regression
from sklearn.linear_model import LinearRegression
```

```python
# Creating the Model
model = LinearRegression()
```

```python
# Fitting the Model
model.fit(X,y)
```

```python
# Predict sales values
predictions = model.predict(X)
print(predictions[:5])
```

# Visualizing the Linear Regression Model

```python
plt.figure(figsize=(8,6))

# Create scatter plot
plt.scatter(X, y, color="green")

# Create line plot
plt.plot(X, predictions, color="red")
plt.xlabel("Radio Expenditure ($)")
plt.ylabel("Sales ($)")

# Display the plot
plt.show()
```
